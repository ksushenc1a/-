```python
# Импорт библиотек
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Задаем переменные для выборок
test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')

# Просмотр тестовой выборки
test.head()
```

```python

# Просмотр тренировочной выборки
train.head()
```

```python
# Проверяем соотношение классов
train.is_joke.value_counts(normalize=True)
```

```python
test.is_joke.value_counts(normalize=True)
```

```python
# Векторизация тренировочной выборки
vectorizer = CountVectorizer()
# В векторайзер нужно засовывать тексты строками (токенизация там встроена)
X = vectorizer.fit_transform(train.text)
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(train.text)
```

```python
X.shape
# Матрица с признаками
# Первая размерность - количество документов
# Вторая размерность - количесто слов в словаре
```

```python
# Нужно задать условия для вхождения токенов (слов) в модель,так как кол-во слов в словаре должно быть либо меньше кол-ва документов, либо быть равным ему
vectorizer = TfidfVectorizer(min_df=5, max_df=0.4)
X = vectorizer.fit_transform(train.text)
X.shape
```

```python
# Векторизация тестовой выборки
vectorizer = TfidfVectorizer(min_df=10, max_df=0.3)
X = vectorizer.fit_transform(train.text)
X_test = vectorizer.transform(test.text)
```

```python
X.shape, X_test.shape
# Матрицы с признаками
# Первая размерность - количество документов
# Вторая размерность - количесто слов в словаре
```

```python
# Векторы с целевой переменной
y = train.is_joke.values
y_test = test.is_joke.values
```

```python
# Логическая регрессия
clf = LogisticRegression(C=0.1, class_weight='balanced')
```

```python
# Обучение модели
clf.fit(X, y)
```

```python
clf.classes_
```

```python
# Предсказание классов
preds = clf.predict(X_test)
```

```python
# Создание датафрейма с результатами
test_df = pd.DataFrame({
    'text': test.iloc[test.index]['text'],
    'actual_class': test.iloc[test.index]['is_joke'],
    'predicted_class': preds
})
test_df
```

```python
# Оценка модели
print(classification_report(y_test, preds))
```

```python
# Байесовский классификатор
clf = MultinomialNB(alpha=1.)
clf.fit(X, y)
preds = clf.predict(X_test)
```

```python
# Создание датафрейма с результатами
test_df = pd.DataFrame({
    'text': test.iloc[test.index]['text'],
    'actual_class': test.iloc[test.index]['is_joke'],
    'predicted_class': preds
})
test_df
test_df.to_csv('bayes.csv', index=False)
```

```python
# Оценка модели
print(classification_report(y_test, preds))
```

```python
# Классификация на основе дерева решений
clf = DecisionTreeClassifier(max_depth=8, class_weight='balanced')
clf.fit(X, y)
preds = clf.predict(X_test)
```

```python
# Создание датафрейма с результатами
test_df = pd.DataFrame({
    'text': test.iloc[test.index]['text'],
    'actual_class': test.iloc[test.index]['is_joke'],
    'predicted_class': preds
})
test_df
```

```python
# Оценка модели
print(classification_report(y_test, preds))
```

```python
# Классификация на основе ближайших соседей
clf = KNeighborsClassifier(n_neighbors=10, metric='cosine')
clf.fit(X, y)
preds = clf.predict(X_test)
```

```python
# Создание датафрейма с результатами
test_df = pd.DataFrame({
    'text': test.iloc[test.index]['text'],
    'actual_class': test.iloc[test.index]['is_joke'],
    'predicted_class': preds
})
test_df
```

```python
# Оценка модели
print(classification_report(y_test, preds))
```

```python
# Классификация на основе множества деревьев решений
rf = RandomForestClassifier(n_estimators=100, max_depth=20)
rf.fit(X, y)
preds = rf.predict(X_test)
```

```python
# Создание датафрейма с результатами
test_df = pd.DataFrame({
    'text': test.iloc[test.index]['text'],
    'actual_class': test.iloc[test.index]['is_joke'],
    'predicted_class': preds
})
test_df
test_df.to_csv('trees.csv', index=False)
```

```python
# Оценка модели
print(classification_report(y_test, preds))
```
